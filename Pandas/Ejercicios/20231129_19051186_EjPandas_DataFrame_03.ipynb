```python
# Importar librer√≠as.
import pandas as pd
import numpy as np

# Leer archivo csv y crear DataFrame.
tips_df = pd.read_csv('c:\\Users\\Admin\\Documents\\ITS\\Semestre 9\\Data_Science\\Pandas\\Datasets\\tips.csv')

# Agregar nueva columna con el porcentaje de la propina.
tips_df["tip_pct"] = tips_df["tip"] / tips_df["total_bill"] * 100
```

```python
tips_df.head()
```

##### 1Ô∏è‚É£ Mover la columna `price_per_person` al final. 

```python
# Obtener una lista de todas las columnas.
cols = list(tips_df.columns)
cols
```

```python
# Eliminar price_per_person de la lista.
cols.remove('price_per_person')
cols
```

```python
# Agregar price_per_person al final de la lista.
cols.append('price_per_person')
```

```python
# Reordenar el DataFrame.
tips_df = tips_df[cols]
tips_df
```

##### 2Ô∏è‚É£ C√≥mo convertir a entero, redondeando hacia arriba.

```python
# Reescribir la columna price_per_person a entero, redondeando hacia arriba.
# El m√©todo ceil() de la librer√≠a numpy redondea hacia arriba.
# El m√©todo astype() de la librer√≠a numpy convierte el valor a entero.
tips_df["price_per_person"] = np.ceil(tips_df["price_per_person"]).astype(int)
```

```python
tips_df
```

##### 3Ô∏è‚É£ Mover la columna `tip_pct` al inicio.

```python
# Obtener una lista de todas las columnas.
cols = list(tips_df.columns)
cols
```

```python
# Eliminar tip_pct de la lista.
cols.remove('tip_pct')
cols
```

```python
# Agregar tip_pct al principio de la lista.
cols.insert(0, 'tip_pct')
cols
```

```python
# Reordenar el DataFrame.
tips_df = tips_df[cols]
```

```python
tips_df
```

##### 4Ô∏è‚É£ Generar nuevo rengl√≥n e insertarlo en el tercer √≠ndice.
