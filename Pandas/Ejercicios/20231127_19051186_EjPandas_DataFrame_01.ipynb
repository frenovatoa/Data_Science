```python
import numpy as np
import pandas as pd
np.random.seed(0)

monedas = ['Peso', 'Dolar', 'Euro', 'Yen', 'Libra']
meses = ['Enero', 'Febrero', 'Marzo', 'Abril']
data = np.random.randint(0, 101, (5, 4))
df = pd.DataFrame(data = data, index = monedas, columns = meses)
```
