Con el dataframe ¬´tips¬ª, realizar lo siguiente:
1. Mostrar la columna ¬´total_bill¬ª en el caso de que la cuenta sea > 20 USD.
2. Mostrar si ¬´total_bill¬ª > 20 USD y ¬´time¬ª es Dinner.
3. Mostrar si ¬´day¬ª es ¬´Sun¬ª o ¬´size¬ª es 3, o ¬´price_per_person¬ª es < 7 USD.
4. Mostrar si ¬´size¬ª es 1 o 2 o 3 con el m√©todo ¬´isin¬ª.
5. 
    - Si el precio de la cuenta es < 10 USD, registrar una nueva columna con el valor de ¬´$¬ª.
    - Si el precio de la cuenta es >= 10 USD y < 30 USD, registrar ¬´$$¬ª.
    - Si el precio de la cuenta es diferente, registrar ¬´$$$¬ª.

```python
import pandas as pd
import numpy as np
tips_df = pd.read_csv('c:\\Users\\Admin\\Documents\\ITS\\Semestre 9\\Data_Science\\Pandas\\Datasets\\tips.csv')
```

#### 1Ô∏è‚É£ Mostrar la columna ¬´total_bill¬ª en el caso de que la cuenta sea > 20 USD.

```python
tips_df[tips_df["total_bill"] > 20]
```

#### 2Ô∏è‚É£ Mostrar si ¬´total_bill¬ª > 20 USD y ¬´time¬ª es Dinner.

```python
tips_df[(tips_df["total_bill"] > 20) & (tips_df["time"] == "Dinner")]
```

#### 3Ô∏è‚É£ Mostrar si ¬´day¬ª es ¬´Sun¬ª o ¬´size¬ª es 3, o ¬´price_per_person¬ª es < 7 USD.

```python
tips_df[(tips_df["day"] == "Sun") | (tips_df["size"] == 3) | (tips_df["price_per_person"] < 7)]
```

#### 4Ô∏è‚É£ Mostrar si ¬´size¬ª es 1 o 2 o 3 con el m√©todo ¬´isin¬ª.

```python
tips_df[tips_df["size"].isin([1, 2, 3])]
```

#### 5Ô∏è‚É£
- Si el precio de la cuenta es < 10 USD, registrar una nueva columna con el valor de ¬´$¬ª.
- Si el precio de la cuenta es >= 10 USD y < 30 USD, registrar ¬´$$¬ª.
- Si el precio de la cuenta es diferente, registrar ¬´$$$¬ª.

```python
tips_df["price_range"] = np.where(tips_df["total_bill"] < 10, "$", 
                                  np.where((tips_df["total_bill"] >= 10) & (tips_df["total_bill"] < 30), "$$", "$$$"))
```
