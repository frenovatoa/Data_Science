Con el dataframe «tips», realizar lo siguiente:
1. Mostrar la columna «total_bill» en el caso de que la cuenta sea > 20 USD.
2. Mostrar si «total_bill» > 20 USD y «time» es Dinner.
3. Mostrar si «day» es «Sun» o «size» es 3, o «price_per_person» es < 7 USD.
4. Mostrar si «size» es 1 o 2 o 3 con el método «isin».
5. 
    - Si el precio de la cuenta es < 10 USD, registrar una nueva columna con el valor de «$».
    - Si el precio de la cuenta es >= 10 USD y < 30 USD, registrar «$$».
    - Si el precio de la cuenta es diferente, registrar «$$$».

```python
import pandas as pd
import numpy as np
tips_df = pd.read_csv('c:\\Users\\Admin\\Documents\\ITS\\Semestre 9\\Data_Science\\Pandas\\Datasets\\tips.csv')
```

#### 1️⃣ Mostrar la columna «total_bill» en el caso de que la cuenta sea > 20 USD.

```python
tips_df[tips_df["total_bill"] > 20]
```

#### 2️⃣ Mostrar si «total_bill» > 20 USD y «time» es Dinner.

```python
tips_df[(tips_df["total_bill"] > 20) & (tips_df["time"] == "Dinner")]
```

#### 3️⃣ Mostrar si «day» es «Sun» o «size» es 3, o «price_per_person» es < 7 USD.

```python
tips_df[(tips_df["day"] == "Sun") | (tips_df["size"] == 3) | (tips_df["price_per_person"] < 7)]
```

#### 4️⃣ Mostrar si «size» es 1 o 2 o 3 con el método «isin».

```python
tips_df[tips_df["size"].isin([1, 2, 3])]
```

#### 5️⃣
- Si el precio de la cuenta es < 10 USD, registrar una nueva columna con el valor de «$».
- Si el precio de la cuenta es >= 10 USD y < 30 USD, registrar «$$».
- Si el precio de la cuenta es diferente, registrar «$$$».

```python
tips_df["price_range"] = np.where(tips_df["total_bill"] < 10, "$", 
                                  np.where((tips_df["total_bill"] >= 10) & (tips_df["total_bill"] < 30), "$$", "$$$"))
```
